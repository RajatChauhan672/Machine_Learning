```python
## Residuals
residuals=y_pred-y_test
```

```python
residuals
```

```python
## Second assumptions
## Distance plot of seaborn which is of kind ="kde"
## And the distance plot should be  a approximate normal distribution

sns.displot(residuals,kind="kde")
```

```python
## Third assumption is that the scatter plot between 
##  Prediction and residuals should be uniform distribution and this is also known as homoscedasticity

plt.scatter(y_pred,residuals)
```
